## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions: 
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

**HCPCS (Healthcare Common Procedure Coding System)** codes are standardised medical billing codes used across Medicare to classify procedures, services, and supplies performed by healthcare providers. In this analysis, we use HCPCS/CPT codes to identify which medical procedures are being performed and whether they are cardiovascular (CVM) in nature.

> **Note:** HCPCS (Healthcare Common Procedure Coding System) / CPT codes are standardised billing codes used across Medicare to classify medical procedures. CVM claims were identified using the top 100 HCPCS/CPT codes associated with cardiovascular procedures.

CVM claims were identified by filtering to the **Top 100 HCPCS/CPT codes** by claim volume across the full Medicare dataset, then isolating codes classified under the **Cardiac** category. This ensures the analysis reflects only the highest-volume, most clinically relevant procedures — consistent with the Medicare Analysis framework discussed in class (Top Procedure Analysis #1).

We identified **6 CVM procedure codes** within the top 100:
- ECG tests: 93010, 93000, 93005
- Echocardiography: 93306
- Myocardial perfusion imaging (SPECT): 78452
- Venous blood collection: 36415

```python
# ─── Filter medicare_df to top-100 HCPCS codes ───────────────────────────────
top100_codes = hcpcs_code_100_df['hcpcs_code'].tolist()

medicare_top100_df = medicare_df[
    medicare_df['hcpcs_code'].isin(top100_codes)
].copy()

# Ensure claim_date is datetime and extract year
medicare_top100_df['claim_date'] = pd.to_datetime(medicare_top100_df['claim_date'])
medicare_top100_df['claim_year'] = medicare_top100_df['claim_date'].dt.year

# Restrict to 2016–2018 only
medicare_top100_df = medicare_top100_df[
    medicare_top100_df['claim_year'].isin([2016, 2017, 2018])
]

print(f'Records in top-100 filtered dataset (2016-2018): {len(medicare_top100_df):,}')
print(f"Years present: {sorted(medicare_top100_df['claim_year'].unique())}")
medicare_top100_df[['claim_id', 'hcpcs_code', 'claim_date', 'claim_year']].head()
```

```python
# ─── Count unique claims by year and CVM status ───────────────────────────────
claims_by_year_cvm = (
    medicare_top100_df
    .groupby(['claim_year', 'is_cvm'])['claim_id']
    .nunique()
    .reset_index()
    .rename(columns={'claim_id': 'unique_claims'})
)

# Map boolean to readable labels
claims_by_year_cvm['claim_type'] = claims_by_year_cvm['is_cvm'].map(
    {True: 'CVM (Cardiac)', False: 'Non-CVM'}
)

print(claims_by_year_cvm)
claims_by_year_cvm
```

```python
# ─── Calculate total claims per year, then compute percentages ───────────────
total_per_year = (
    claims_by_year_cvm
    .groupby('claim_year')['unique_claims']
    .sum()
    .rename('total_claims')
)

claims_by_year_cvm = claims_by_year_cvm.merge(total_per_year, on='claim_year')
claims_by_year_cvm['pct_share'] = (
    claims_by_year_cvm['unique_claims'] / claims_by_year_cvm['total_claims'] * 100
)

print(claims_by_year_cvm[['claim_year', 'claim_type', 'unique_claims', 'total_claims', 'pct_share']])
```

```python
# ─── Count unique CVM claims per HCP (across all years combined) ─────────────
hcp_cvm_volume_df = (
    cvm_claims_df
    .groupby('npi_id')['claim_id']
    .nunique()
    .reset_index()
    .rename(columns={'claim_id': 'total_cvm_claims'})
    .sort_values('total_cvm_claims', ascending=False)
    .reset_index(drop=True)
)

print(f'Total unique HCPs with CVM claims: {len(hcp_cvm_volume_df):,}')
print()
print('Distribution summary:')
print(hcp_cvm_volume_df['total_cvm_claims'].describe().round(2))
hcp_cvm_volume_df.head(10)
```

```python
# ─── Define tier bins and labels ─────────────────────────────────────────────
bins   = [0, 1, 5, 10, 25, float('inf')]
labels = ['Tier 1: 1 claim\n(Very Low)',
          'Tier 2: 2-5 claims\n(Low)',
          'Tier 3: 6-10 claims\n(Moderate)',
          'Tier 4: 11-25 claims\n(High)',
          'Tier 5: 25+ claims\n(Very High / KOL)']

hcp_cvm_volume_df['tier'] = pd.cut(
    hcp_cvm_volume_df['total_cvm_claims'],
    bins=bins,
    labels=labels,
    right=True
)

# ─── Aggregate: HCP count and % per tier ──────────────────────────────────────
tier_summary_df = (
    hcp_cvm_volume_df
    .groupby('tier', observed=True)
    .agg(
        hcp_count=('npi_id', 'count'),
        total_claims_in_tier=('total_cvm_claims', 'sum'),
        avg_claims_per_hcp=('total_cvm_claims', 'mean')
    )
    .reset_index()
)

total_hcps = tier_summary_df['hcp_count'].sum()
total_claims = tier_summary_df['total_claims_in_tier'].sum()

tier_summary_df['pct_of_hcps']   = (tier_summary_df['hcp_count'] / total_hcps * 100).round(1)
tier_summary_df['pct_of_claims'] = (tier_summary_df['total_claims_in_tier'] / total_claims * 100).round(1)
tier_summary_df['avg_claims_per_hcp'] = tier_summary_df['avg_claims_per_hcp'].round(1)

tier_summary_df
```

```python
import seaborn as sns


cvm_df = cvm_claims_df.copy()
cvm_df = cvm_df[cvm_df['patient_age'].between(0, 120)]

# ── Step 2: Assign required age groups ───────────────────────────────────────
age_bins   = [17, 59, 69, 79, float('inf')]
age_labels = ['18–59', '60–69', '70–79', '80+']

cvm_df['age_group'] = pd.cut(
    cvm_df['patient_age'],
    bins=age_bins,
    labels=age_labels,
    right=True
)

# ── Step 3: Assign HCP tiers based on total CVM claims per provider ───────────
# Reuse hcp_cvm_volume_df already computed in Q2
# Tier assignment based on 3-year total claim count per NPI

def assign_hcp_tier(n):
    if n == 1:
        return 'Tier 1: 1 claim'
    elif 2 <= n <= 4:
        return 'Tier 2: 2–4 claims'
    elif 5 <= n <= 9:
        return 'Tier 3: 5–9 claims'
    else:
        return 'Tier 4: 10+ claims'

hcp_tier_map = hcp_cvm_volume_df[['npi_id', 'total_cvm_claims']].copy()
hcp_tier_map['hcp_tier'] = hcp_tier_map['total_cvm_claims'].apply(assign_hcp_tier)

# Merge tier into working dataframe
cvm_df = cvm_df.merge(
    hcp_tier_map[['npi_id', 'hcp_tier']],
    on='npi_id',
    how='left'
)

tier_order = ['Tier 1: 1 claim', 'Tier 2: 2–4 claims',
              'Tier 3: 5–9 claims', 'Tier 4: 10+ claims']

print(f"Records after merge: {len(cvm_df):,}")
print(f"Age groups: {cvm_df['age_group'].value_counts().to_dict()}")
print(f"HCP tiers : {cvm_df['hcp_tier'].value_counts().to_dict()}")

# ─────────────────────────────────────────────────────────────────────────────
# OUTPUT 1: Pivot table — unique claims by age group × HCP tier
# ─────────────────────────────────────────────────────────────────────────────

pivot_df = (
    cvm_df
    .groupby(['age_group', 'hcp_tier'], observed=True)['claim_id']
    .nunique()
    .reset_index()
    .rename(columns={'claim_id': 'unique_claims'})
    .pivot(index='age_group', columns='hcp_tier', values='unique_claims')
    .reindex(index=age_labels, columns=tier_order)
    .fillna(0)
    .astype(int)
)

# Add totals
pivot_df['Total'] = pivot_df.sum(axis=1)
pivot_df.loc['Total'] = pivot_df.sum()

print("\nCVM Claims Distribution by Age Group and HCP Tier:")
display(pivot_df)

# ─────────────────────────────────────────────────────────────────────────────
# OUTPUT 2: Heatmap — claims concentration across age × tier
# ─────────────────────────────────────────────────────────────────────────────

# Exclude totals row/col for the heatmap
heatmap_data = pivot_df.drop(index='Total', columns='Total')

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#FAFCFF')
ax.set_facecolor('#FAFCFF')

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='d',
    cmap='Blues',
    linewidths=0.5,
    linecolor='white',
    ax=ax,
    annot_kws={'size': 11, 'weight': 'bold', 'color': '#1A3A5C'},
    cbar_kws={'label': 'Unique CVM Claims'}
)

ax.set_title(
    'CVM Claims Distribution by Age Group and HCP Tier',
    fontsize=14, fontweight='bold', color='#1A3A5C', pad=16
)
ax.set_xlabel('HCP Tier (by total CVM claim volume)', fontsize=11, color='#2C3E50', labelpad=8)
ax.set_ylabel('Patient Age Group', fontsize=11, color='#2C3E50', labelpad=8)
ax.tick_params(axis='both', colors='#2C3E50', labelsize=10)
ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.tight_layout()
plt.savefig('Q4_Heatmap_Age_HCP_Tier.png', dpi=150,
            bbox_inches='tight', facecolor='#FAFCFF')
plt.show()
print("Saved as Q4_Heatmap_Age_HCP_Tier.png")

# ─────────────────────────────────────────────────────────────────────────────
# OUTPUT 3: YoY growth % per age group (2016→2017 and 2017→2018)
# ─────────────────────────────────────────────────────────────────────────────

# Count unique claims per age group per year
age_year_df = (
    cvm_df
    .groupby(['claim_year', 'age_group'], observed=True)['claim_id']
    .nunique()
    .reset_index()
    .rename(columns={'claim_id': 'claims'})
)

age_year_pivot = age_year_df.pivot(
    index='claim_year', columns='age_group', values='claims'
).fillna(0).astype(int)[age_labels]

# YoY % change
yoy = (age_year_pivot.pct_change() * 100).round(1).iloc[1:]
yoy.index = ['2016→2017', '2017→2018']

# Average YoY across both periods — used for scatter plot
avg_yoy = yoy.mean().round(1)

# Total claims per age group (2018, most recent year)
total_2018 = age_year_pivot.loc[2018]

print("\nYoY % Change in CVM Claims by Age Group:")
display(
    yoy.style
    .format("{:+.1f}%")
    .applymap(lambda v: 'color:#2e7d32;font-weight:bold' if v > 0
              else 'color:#c62828;font-weight:bold')
    .set_caption("Green = growth | Red = decline")
)


# ─────────────────────────────────────────────────────────────────────────────
# OUTPUT 4: Scatter plot — YoY growth vs total claims (underpenetration map)
# ─────────────────────────────────────────────────────────────────────────────

scatter_colors = ['#90BEE0', '#4A90C4', '#1A5A8A', '#0D3A5C']

fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#FAFCFF')
ax.set_facecolor('#FAFCFF')

mean_x = avg_yoy.mean()
mean_y = total_2018.mean()

# ── Shaded quadrant backgrounds ───────────────────────────────────────────────
x_min, x_max = avg_yoy.min() - 4, avg_yoy.max() + 6
y_min, y_max = total_2018.min() - 200, total_2018.max() + 300

ax.fill_betweenx([mean_y, y_max], mean_x, x_max,
                 color='#E8F4E8', alpha=0.5, zorder=0)   # top-right: green tint
ax.fill_betweenx([y_min, mean_y], mean_x, x_max,
                 color='#E8F0F8', alpha=0.4, zorder=0)   # bottom-right: blue tint
ax.fill_betweenx([mean_y, y_max], x_min, mean_x,
                 color='#F5F5F5', alpha=0.5, zorder=0)   # top-left: grey
ax.fill_betweenx([y_min, mean_y], x_min, mean_x,
                 color='#FFF5F5', alpha=0.4, zorder=0)   # bottom-left: faint red

# ── Reference lines ───────────────────────────────────────────────────────────
ax.axvline(mean_x, color='#A0B8CC', linestyle='--', linewidth=1.5, zorder=1)
ax.axhline(mean_y, color='#A0B8CC', linestyle='--', linewidth=1.5, zorder=1)

# ── Quadrant labels (corners) ─────────────────────────────────────────────────
quad_style = dict(fontsize=10, alpha=0.75, style='italic')

ax.text(mean_x + 0.5, y_max - 60,
        'HIGH GROWTH · HIGH VOLUME\nPrimary investment zone',
        color='#2e7d32', ha='left', **quad_style)

ax.text(x_min + 0.3, y_max - 60,
        'LOW GROWTH · HIGH VOLUME\nDefend & retain',
        color='#555555', ha='left', **quad_style)

ax.text(mean_x + 0.5, y_min + 60,
        'HIGH GROWTH · LOW VOLUME\nEmerging — invest early',
        color='#1A5A8A', ha='left', **quad_style)

ax.text(x_min + 0.3, y_min + 60,
        'LOW GROWTH · LOW VOLUME\nDeprioritise',
        color='#999999', ha='left', **quad_style)

# ── Count unique providers per age group (2018) ───────────────────────────────
# Bubble size = provider count — shows where demand outpaces provider coverage
provider_count_2018 = (
    cvm_df[cvm_df['claim_year'] == 2018]
    .groupby('age_group', observed=True)['npi_id']
    .nunique()
    .reindex(age_labels)
    .fillna(0)
    .astype(int)
)
size_scale = 0.4

label_offsets = {
    '18–59': (1.2,  -120),
    '60–69': (-6,    120),
    '70–79': (0.8,   120),
    '80+':   (1.2,  -130),
}

for i, age_grp in enumerate(age_labels):
    x = avg_yoy[age_grp]
    y = total_2018[age_grp]
    size = size = provider_count_2018[age_grp] * size_scale

    ax.scatter(x, y,
               s=size,
               color=scatter_colors[i],
               zorder=5,
               edgecolors='white',
               linewidths=2.5,
               alpha=0.92)

    dx, dy = label_offsets[age_grp]

    ax.annotate(
        f'{age_grp}\n{int(y):,} claims\n{int(provider_count_2018[age_grp]):,} providers\nAvg YoY: +{x:.1f}%',
        xy=(x, y),
        xytext=(x + dx, y + dy),
        fontsize=10,
        fontweight='bold',
        color=scatter_colors[i],
        ha='center',
        bbox=dict(
            boxstyle='round,pad=0.35',
            facecolor='white',
            edgecolor=scatter_colors[i],
            linewidth=1.2,
            alpha=0.95
        ),
        arrowprops=dict(
            arrowstyle='-',
            color=scatter_colors[i],
            lw=1.2,
            shrinkA=0,
            shrinkB=6
        )
    )

# ── Axes ──────────────────────────────────────────────────────────────────────
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
ax.set_xlabel('Average YoY Growth Rate (%) — 2016 to 2018',
              fontsize=12, color='#2C3E50', labelpad=10)
ax.set_ylabel('Total CVM Claims (2018)',
              fontsize=12, color='#2C3E50', labelpad=10)
ax.set_title(
    'CVM Opportunity Map: Claim Volume vs Growth Rate by Patient Age Segment\n(Bubble size = number of unique providers serving each segment in 2018)',
     fontsize=13, fontweight='bold', color='#1A3A5C', pad=18
)
ax.tick_params(colors='#2C3E50', labelsize=10)
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.grid(axis='both', linestyle='--', alpha=0.3, color='#C8D8E8')
ax.spines[['top', 'right']].set_visible(False)
ax.spines['left'].set_color('#C8D8E8')
ax.spines['bottom'].set_color('#C8D8E8')

# ── Bubble size legend ────────────────────────────────────────────────────────
for vol, lbl in [(500, '500 providers'), (1000, '1,000 providers'), (1500, '1,500 providers')]:
    ax.scatter([], [], s=vol * size_scale, color='#90BEE0',
               alpha=0.7, edgecolors='white', linewidths=1.5, label=lbl)
ax.legend(title='Bubble size = provider count (2018)',
          loc='lower right', fontsize=9, title_fontsize=9,
          frameon=True, framealpha=0.9, edgecolor='#C8D8E8')

plt.tight_layout()
plt.savefig('Q4_Scatter_Underpenetration.png', dpi=150,
            bbox_inches='tight', facecolor='#FAFCFF')
plt.show()
print("Saved as Q4_Scatter_Underpenetration.png")
```

###  Executive Summary: Underpenetrated CVM Patient–Provider Segments

**Business Question:**
Where in the CVM landscape is strong patient demand not being matched
by proportional high-tier provider engagement — and what does this
mean for sales force deployment?

**What we analysed:**
We mapped 12,425 unique CVM claims (2016–2018) across 4 patient age
segments (18–59, 60–69, 70–79, 80+) and 4 HCP engagement tiers
(Tier 1: 1 claim → Tier 4: 10+ claims). We then overlaid 2-year
YoY growth rates and 2018 provider counts to identify where high
patient demand is not proportionally served by high-tier providers.


**Note on heatmap metric:** The heatmap displays unique CVM claim
counts (not provider counts) by age group and HCP tier. This is
intentional — claims represent the actual volume of cardiovascular
care being delivered in each segment, which is the true measure of
commercial opportunity. Provider counts are used separately in the
Opportunity Map scatter plot to show coverage relative to demand.


---

**Key Finding 1 — Tier 1 dominance signals systemic underpenetration**

Across every age group, Tier 1 providers (1 claim/year) account for
the majority of CVM claims:
- 70–79 segment: 2,216 claims from Tier 1 vs only 141 from Tier 3
- 60–69 segment: 2,038 claims from Tier 1 vs only 106 from Tier 3
- 80+ segment:   1,636 claims from Tier 1 vs only 134 from Tier 3

This means the bulk of CVM patient management is happening with
providers who are barely engaged. Tier 3 HCPs (5–9 claims) —
who have demonstrated sustained CVM intent — are severely
underrepresented across all segments.

**Key Finding 2 — 70–79 is the primary underpenetrated opportunity**

The 70–79 segment is in the HIGH GROWTH · HIGH VOLUME quadrant:
- 1,757 CVM claims in 2018 (highest of all segments)
- 1,237 unique providers (largest provider base)
- +29.8% average YoY growth (fastest growing segment)

Despite this, the heatmap confirms Tier 1 providers handle 2,216
claims vs Tier 3's 141 — a 16:1 ratio. This segment has both
the demand and the provider base to support rapid tier uplift.

**Key Finding 3 — 80+ is the surprise high-growth segment**

The 80+ segment delivered +37.4% YoY growth in 2018 — the highest
of any group — yet sits just below the volume mean. With 919
providers and growing patient demand, this segment is transitioning
from emerging to core and requires immediate investment before
competitors establish provider relationships first.

**Key Finding 4 — 60–69 is stable, not growing**

60–69 sits in the LOW GROWTH · HIGH VOLUME quadrant (+3.9% avg
YoY) — this is a mature segment to defend, not expand. Reduce
incremental field force investment here and maintain through
cost-efficient NPP channels.

---

**Strategic Recommendations:**

| Priority | Segment | Action | Channel |
|---|---|---|---|
| 1 — Immediate | 70–79 | Convert Tier 1 → Tier 3 HCPs | Field force rep visits + peer education |
| 2 — Urgent | 80+ | Build provider relationships before market matures | Medical affairs + KOL programs |
| 3 — Pipeline | 18–59 | Early digital engagement with specialty HCPs | NPP — digital, email, specialist networks |
| 4 — Maintain | 60–69 | Defend existing relationships cost-efficiently | Reduce field force, increase NPP |

**The single most impactful action:**
Target the ~2,216 Tier 1 HCPs in the 70–79 segment with a structured
conversion program. Moving just 15% of them to Tier 3 would add
~330 high-engagement providers to the CVM pipeline — equivalent to
doubling the current Tier 3 provider base across all age groups.
